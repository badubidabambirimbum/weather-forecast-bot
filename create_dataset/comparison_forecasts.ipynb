```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Hourly
from datetime import datetime
from geopy.geocoders import Nominatim
import os
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_Msc_Ya = pd.read_csv('../data/Moscow_Yandex_10.csv', sep=',', index_col='date')
df_Msc_Ya
```

```python
df_Msc_Gis = pd.read_csv('../data/Moscow_GisMeteo_10.csv', sep=',', index_col='date')
df_Msc_Gis
```

```python
full_date_range = pd.date_range(start=df_Msc_Gis.index.min(), end=df_Msc_Gis.index.max(), freq='D')
missing_dates = full_date_range.difference(df_Msc_Gis.index)
missing_dates
```

```python
def get_city_coordinates(city_name):
    geolocator = Nominatim(user_agent="weather_data_fetcher")
    location = geolocator.geocode(city_name)
    return location.latitude, location.longitude

city_name = "Moscow"
latitude, longitude = get_city_coordinates(city_name)

start = datetime(2024, 5, 15)
end = datetime(2024, 10, 8)


location = Point(latitude, longitude)

df2 = Daily(location, start, end)
df2 = df2.fetch()

df2.drop(index=['2024-06-27', '2024-06-28', '2024-06-29', '2024-08-04'], inplace=True)
df2
```

```python
df_night = df2.iloc[:,1].values
df_night_real = []

for i in range(len(df_night)-10+1):
    df_night_real.append(df_night[i:i+10])

df_night_real = np.array(df_night_real)
df_night_real
```

```python
df_day = df2.iloc[:,2].values
df_day_real = []

for i in range(len(df_day)-10+1):
    df_day_real.append(df_day[i:i+10])

df_day_real = np.array(df_day_real)
df_day_real
```

```python
df_msc_ya_day = df_Msc_Ya.iloc[:,:10].values
df_msc_ya_day
```

```python
df_msc_ya_night = df_Msc_Ya.iloc[:,10:20].values
df_msc_ya_night
```

```python
df_msc_gis_day = df_Msc_Gis.iloc[:,:10].values
df_msc_gis_day
```

```python
df_msc_gis_night = df_Msc_Gis.iloc[:,10:20].values
df_msc_gis_night
```

```python
list_mae_day_ya = []
list_mae_day_gis = []

for i in range(10):
    mae_day_ya = np.mean(np.abs(df_day_real[:, i] - df_msc_ya_day[:, i]))
    mae_day_gis = np.mean(np.abs(df_day_real[:, i] - df_msc_gis_day[:, i]))

    list_mae_day_ya.append(mae_day_ya)
    list_mae_day_gis.append(mae_day_gis)

    print(f'MAE for Day Yandex {i+1:<2}: {mae_day_ya:<6.4f} \t\t| MAE for Day GisMeteo {i+1:<2}: {mae_day_gis:<6.4f}')

print(f"Day Yandex: {np.mean(np.array(list_mae_day_ya)):<6.4f} \t\t\t| Day GisMeteo: {np.mean(np.array(list_mae_day_gis)):<6.4f}")
```

```python
list_mae_night_ya = []
list_mae_night_gis = []

for i in range(10):
    mae_night_ya = np.mean(np.abs(df_night_real[:, i] - df_msc_ya_night[:, i]))
    mae_night_gis = np.mean(np.abs(df_night_real[:, i] - df_msc_gis_night[:, i]))

    list_mae_night_ya.append(mae_night_ya)
    list_mae_night_gis.append(mae_night_gis)

    print(f'MAE for Night Yandex {i+1:<2}: {mae_night_ya:<6.4f} \t| MAE for Night GisMeteo {i+1:<2}: {mae_night_gis:<6.4f}')

print(f"Night Yandex: {np.mean(np.array(list_mae_night_ya)):<6.4f} \t\t\t| Night GisMeteo: {np.mean(np.array(list_mae_night_gis)):<6.4f}")
```
